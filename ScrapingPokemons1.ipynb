```python
import pandas as pd
import requests

# URL za pridobivanje podatkov o Pokémonih
pokemon_api_url = "https://pokeapi.co/api/v2/pokemon/"

# Število želenih instanc Pokémonov
stevilo_instanc = 1000

# Seznam za shranjevanje podatkov
pokemon_podatki = []

# Pridobivanje podatkov o vsakem Pokémonu
for pokemon_id in range(1, stevilo_instanc + 1):
    url = f"{pokemon_api_url}{pokemon_id}"
    odgovor = requests.get(url)

    if odgovor.status_code == 200:
        podatki = odgovor.json()
        pokemon_podatki.append({
            "Ime": podatki["name"],
            "Teza": podatki["weight"],
            "Visina": podatki["height"],
            "Tip": podatki["types"][0]["type"]["name"],
            "Skupna osnovna statistika": podatki["base_experience"],
            "Tip2": podatki["types"][1]["type"]["name"] if len(podatki["types"]) > 1 else None,  # Dodana druga vrsta, če obstaja
            "Sposobnost": podatki["abilities"][0]["ability"]["name"],  # Prva sposobnost
            "HP": podatki["stats"][0]["base_stat"],
            "Napad": podatki["stats"][1]["base_stat"],
            "Obramba": podatki["stats"][2]["base_stat"],
            "Hitrost": podatki["stats"][-1]["base_stat"],
            "PosebniNapad": podatki["stats"][3]["base_stat"],
            "PosebnaObramba": podatki["stats"][4]["base_stat"],
        })
    else:
        print(f"Napaka pri pridobivanju podatkov za Pokémona {pokemon_id}")

# Pretvorba seznama v pandas DataFrame
pokemon_df = pd.DataFrame(pokemon_podatki)

# Shranjevanje podatkov v CSV datoteko
pokemon_df.to_csv("pokemon_podatki.csv", index=False)

# Izpis nekaj vrstic podatkov
print(pokemon_df.head())
```
